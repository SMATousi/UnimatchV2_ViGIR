```python
import pandas as pd

# Load the CSV file
file_path = "./wandb_export_2025-02-05T10 03 55.910-06 00.csv"
df = pd.read_csv(file_path)

# Display the first few rows to understand its structure
df.head()
```

```python
import matplotlib.pyplot as plt
import re

# Filter the rows that contain "wo_FA" but not "FA" (to avoid capturing "FA" alone)
wo_FA_df = df[df['Name'].apply(lambda x: 'wo_FA' in x and 'FA' in x and 'FA' not in x.replace('wo_FA', ''))]

# Extract the split numbers or labels from the names
wo_FA_df['Split'] = wo_FA_df['Name'].apply(lambda x: re.search(r'(\d+|full)', x).group(0) if re.search(r'(\d+|full)', x) else None)

# Convert numeric splits to integers for proper sorting
wo_FA_df['Split'] = wo_FA_df['Split'].apply(lambda x: int(x) if x.isdigit() else x)

# Sort by split values
wo_FA_df = wo_FA_df.sort_values(by='Split', key=lambda x: x.map(lambda v: float('inf') if v == 'full' else v))

strict_wo_FA_df = df[df['Name'].apply(lambda x: bool(re.fullmatch(r'pascal_wo_FA_(\d+|full)', x)))]

# Extract the split numbers or labels
strict_wo_FA_df['Split'] = strict_wo_FA_df['Name'].apply(lambda x: re.search(r'(\d+|full)', x).group(0))

# Convert numeric splits to integers for proper sorting, map 'full' to 100
strict_wo_FA_df['Split'] = strict_wo_FA_df['Split'].apply(lambda x: 100 if x == 'full' else int(x))

# Sort by split values
strict_wo_FA_df = strict_wo_FA_df.sort_values(by='Split')


strict_wo_FA_df['Split'] = strict_wo_FA_df['Split'].replace(100, 1464)

# Sort again after modification
strict_wo_FA_df = strict_wo_FA_df.sort_values(by='Split')

# Plot eval/mIoU_ema (Max) vs Split
plt.figure(figsize=(10, 5))
plt.plot(strict_wo_FA_df['Split'], strict_wo_FA_df['eval/mIoU (Max)'], marker='o', linestyle='-', label="wo_FA")

plt.xlabel('Split')
plt.ylabel('eval/mIoU_ema (Max)')
plt.title('Evaluation IoU EMA for pascal_wo_FA across Splits')
plt.xticks(strict_wo_FA_df['Split'], labels=strict_wo_FA_df['Split'].astype(str), rotation=45)
plt.grid(True)
plt.legend()
plt.show()

```

```python
all_run_names = [r'pascal_wo_FA_(\d+|full)_norm',
                 r'pascal_wo_FA_(\d+|full)_gradient',
                r'pascal_wo_FA_(\d+|full)_norm_gradient',
                r'pascal_FA_(\d+|full)',
                r'pascal_FA_(\d+|full)_norm',
                r'pascal_FA_(\d+|full)_norm_gradient',
                r'pascal_FA_(\d+|full)_gradient']

for run_name in all_run_names:
    
    # strict_FA_df = df[df['Name'].apply(lambda x: bool(re.fullmatch(r'pascal_FA_(\d+|full)', x)))]
    strict_FA_df = df[df['Name'].apply(lambda x: bool(re.fullmatch(run_name, x)))]
    
    
    # Extract the split numbers or labels
    strict_FA_df['Split'] = strict_FA_df['Name'].apply(lambda x: re.search(r'(\d+|full)', x).group(0))
    
    # Convert numeric splits to integers for proper sorting, map 'full' to 1464
    strict_FA_df['Split'] = strict_FA_df['Split'].apply(lambda x: 1464 if x == 'full' else int(x))
    
    # Sort by split values
    strict_FA_df = strict_FA_df.sort_values(by='Split')
    
    # Merge the two dataframes to compute the improvement
    comparison_df = pd.merge(strict_wo_FA_df[['Split', 'eval/mIoU_ema (Max)']],
                             strict_FA_df[['Split', 'eval/mIoU_ema (Max)']],
                             on='Split', suffixes=('_wo_FA', '_FA'))
    
    # Compute the improvement
    comparison_df['Improvement'] = comparison_df['eval/mIoU_ema (Max)_FA'] - comparison_df['eval/mIoU_ema (Max)_wo_FA']
    
    # Plot the improvements
    plt.figure(figsize=(10, 5))
    plt.plot(comparison_df['Split'], comparison_df['Improvement'], marker='o', linestyle='-', color='r', label="Improvement")
    
    plt.xlabel('Split')
    plt.ylabel('Improvement in eval/mIoU_ema (Max)')
    plt.title(run_name)
    plt.xticks(comparison_df['Split'], labels=comparison_df['Split'].astype(str), rotation=45)
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
import re
import pandas as pd
import matplotlib.pyplot as plt

all_run_names = [r'pascal_wo_FA_(\d+|full)_norm',
                 r'pascal_wo_FA_(\d+|full)_gradient',
                 r'pascal_wo_FA_(\d+|full)_norm_gradient',
                 r'pascal_FA_(\d+|full)',
                 r'pascal_FA_(\d+|full)_norm',
                 r'pascal_FA_(\d+|full)_norm_gradient',
                 r'pascal_FA_(\d+|full)_gradient']

plt.figure(figsize=(10, 6))  # Create a single figure for all plots

for run_name in all_run_names:
    strict_FA_df = df[df['Name'].apply(lambda x: bool(re.fullmatch(run_name, x)))]

    # Extract and process the 'Split' column
    strict_FA_df['Split'] = strict_FA_df['Name'].apply(lambda x: re.search(r'(\d+|full)', x).group(0))
    strict_FA_df['Split'] = strict_FA_df['Split'].apply(lambda x: 1464 if x == 'full' else int(x))
    strict_FA_df = strict_FA_df.sort_values(by='Split')

    # Merge with `strict_wo_FA_df` for improvement calculation
    comparison_df = pd.merge(strict_wo_FA_df[['Split', 'eval/mIoU (Max)']],
                             strict_FA_df[['Split', 'eval/mIoU (Max)']],
                             on='Split', suffixes=('_wo_FA', '_FA'))

    # Compute the improvement
    comparison_df['Improvement'] = comparison_df['eval/mIoU (Max)_FA'] - comparison_df['eval/mIoU (Max)_wo_FA']

    # Plot all lines on the same figure
    plt.plot(comparison_df['Split'], comparison_df['Improvement'], marker='o', linestyle='-', label=run_name.replace('_(\d+|full)',''))

# Final plot settings
plt.xlabel('Split')
plt.ylabel('Improvement in eval/mIoU_ema (Max)')
plt.title('Improvement Comparison for Different Runs')
plt.xticks(comparison_df['Split'], labels=comparison_df['Split'].astype(str), rotation=45)
plt.grid(True)
plt.legend()
plt.show()
```
