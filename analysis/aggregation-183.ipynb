```python
import pandas as pd

# Load the CSV file to examine the data
file_path = './wandb-183.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
data.head()
```

```python
# Group by the 'Name' column and average the 'eval/mIoU_ema' values for the names that are the same
average_mIoU_ema = data.groupby('Name')['eval/mIoU_ema (Max)'].mean().reset_index()

# Show the results
average_mIoU_ema
```

```python
# Sort the DataFrame in descending order by the 'eval/mIoU_ema' column
sorted_average_mIoU_ema = average_mIoU_ema.sort_values(by='eval/mIoU_ema (Max)', ascending=False).reset_index(drop=True)

# Show the sorted results
sorted_average_mIoU_ema
```

```python
# Split the sorted DataFrame into three groups based on the size mentioned in the name: 'small', 'base', and 'large'
small_models = sorted_average_mIoU_ema[sorted_average_mIoU_ema['Name'].str.contains('small')]
base_models = sorted_average_mIoU_ema[sorted_average_mIoU_ema['Name'].str.contains('base')]
large_models = sorted_average_mIoU_ema[sorted_average_mIoU_ema['Name'].str.contains('large')]

# Show the separated groups
small_models, base_models, large_models
```
